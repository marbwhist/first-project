```python
# Import dependencies.
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import seaborn as sns
from scipy.stats import linregress
from pandas.io.json import json_normalize
import json
from collections import Counter
from wordcloud import WordCloud, STOPWORDS
```

```python
# Import our new csv to analyze and convert to a dataframe.
movies = "movie_data.csv"
movie_data = pd.read_csv(movies)
movie_data.head()
```

```python
# Make our new dataframe to start analyzing synopsis
movie_df = movie_data.drop(movie_data.columns[[ 0,3, 5,6,7,10, 11,12]], axis=1)
movie_df
```

```python
# Create a list to hold words
master_list=[]

# Create a for loop to run a split on the string in the synopsis column
for index, row in movie_df.iterrows():

             
    themes=str(row['synopsis']).split(' ')
    
    # Adding the synopsis words into the new Master List
    master_list.append(themes)
    
# Print list to check our work
print(master_list)


# Create a list to count words
word_counter=[]

# Create a loop to run a new list within the Master List and count the most recurring words
for li in master_list:
    word_count=Counter(li).values()

    word_count_key=Counter(li).keys()
    most_occur=Counter(li).most_common(5)
    print(most_occur)


    # Zip together the word count key and word count list, and add the words to our word counter list
    word_dictionary=dict(zip(word_count_key, word_count))
    word_counter.append(word_dictionary)

# Make a new dataframe for our word counter
word_df=pd.DataFrame(word_counter)

# Print the dataframe
print(word_df)
```

```python
for val in most_occur: 
      
    # typecaste each val to string 
    val = str(val) 
  
    # split the value 
    tokens = val.split() 
      
# Converts each token into lowercase 
for i in range(len(tokens)): 
     tokens[i] = tokens[i].lower() 
      
# Create our word cloud    
cloud = WordCloud(width = 800, height = 800, 
               background_color ='white').generate((' '.join(sum(master_list, []))))
 
  
# plot the WordCloud image                        
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(cloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
plt.title('Most Frequently Used Words in Film Synopsis', fontsize=30)
plt.show()
plt.savefig('plot_word_cloud.png')
```

```python
# Set up a new dataframe with focused data for our cast analysis
movie_df = movie_data.drop(["Genre_2", "Genre_3","synopsis"], axis=1)
movie_df
```

```python
# Run analysis on worldwide gross (Earnings)
print(movie_df["worldwide_gross"].mean())
print(movie_df["worldwide_gross"].median())
print(movie_df["worldwide_gross"].max())
print(movie_df["worldwide_gross"].min())
```

```python
# used if function to only run if it is not already a list

if type(movie_df["cast"][0]) != list:    
    movie_df["cast"] = movie_df["cast"].str.split(',').tolist()
```

```python
# used explode function to separate each actor

cast_df = movie_df.explode("cast")
cast_df.head()
```

```python
# created summary of average sales, IMDB rating, and metacritc for the the films each actor was in

actor_averages = cast_df.groupby(["cast"])
# actor_averages.head()
actor_gross = ((actor_averages["worldwide_gross"].mean()) / 1000000).round()

actor_imdb_rating = ((actor_averages["imdb_rating"].mean()) * 10).round()

actor_metacritic = (actor_averages["metacritic"].mean()).round()

actor_films = actor_averages["title"].count()

films = actor_averages["title"]

actor_stats = pd.DataFrame({"Average Worldwide Gross Sales": actor_gross,
                             "Average IMDB Rating": actor_imdb_rating,
                             "Average Metacritic": actor_metacritic,
                             "Number of Films": actor_films,
                             "Films": films})

actor_stats
```

```python
# show the table with actors who were in more than 1 movie

top_actors = actor_stats.loc[actor_stats["Number of Films"] > 1, [
    "Average Worldwide Gross Sales", "Average IMDB Rating", "Average Metacritic", "Number of Films", "Films"]]
top_actors
```

```python

# showed the top ten high grossing actors with more than 1 movie

top_actor_stats = top_actors.loc[top_actors["Average Worldwide Gross Sales"] > 976, [
    "Average Worldwide Gross Sales", "Average IMDB Rating", "Average Metacritic", "Number of Films", "Films"]]
```

```python
# sorted table

sort_actors = top_actor_stats.sort_values(by=["Average Worldwide Gross Sales"], ascending=False)

sort_actors
```

```python
x_axis = np.arange(len(sort_actors))
tick_locations = [value for value in x_axis]

plt.figure(figsize=(20,3))
plt.bar(x_axis, sort_actors["Average Worldwide Gross Sales"], color='b', alpha=0.5)
plt.xticks(tick_locations, ["Zoe Saldana", "Michael Gambon", "Christian Bale", "Johnny Depp", "Josh Duhamel", "Shia LaBouf", "Ken Stott", "Martin Freeman", "Richard Armitage", "Ian McKellen"])
plt.title("Average Gross Sales per Actor")
plt.ylabel("Gross Sales")
# Save as PNG
plt.savefig('avg_sales_per_actor.png')
```

```python
# Created tables for averages by studio

studio_averages = movie_df.groupby(["studio"])

average_gross = ((studio_averages["worldwide_gross"].mean()) / 1000000).round()

average_imdb_rating = ((studio_averages["imdb_rating"].mean()) * 10).round()

average_metacritic = (studio_averages["metacritic"].mean()).round()

number_films = studio_averages["title"].count()

studio_stats = pd.DataFrame({"Average Worldwide Gross Sales": average_gross,
                             "Average IMDB Rating": average_imdb_rating,
                             "Average Metacritic": average_metacritic,
                             "Number of Films": number_films})

studio_stats
```

```python
# bar chart for average gross sales by studio

x_axis = np.arange(len(studio_stats))
tick_locations = [value for value in x_axis]

plt.figure(figsize=(20,3))
plt.bar(x_axis, studio_stats["Average Worldwide Gross Sales"], color='b', alpha=0.5, align="center")
plt.xticks(tick_locations, ["20th Century Fox", "Columbia Pictures", "Lionsgate Films", "Metro-Goldwyn-Mayer", "Paramount Pictures","Pixar", "Sony Pictures", "Universal Pictures", "Walt Disney Pictures", "Warner Bros"])
plt.title("Average Gross Sales per Studio")
plt.ylabel("Gross Sales")
# Save as PNG
plt.savefig('avg_sales_per_studio.png')
```

```python
# Created tables for averages by title

title_averages = movie_df.groupby(["title"])

title_gross = ((title_averages["worldwide_gross"].mean()) / 1000000).round()

title_imdb_rating = ((title_averages["imdb_rating"].mean()) * 10).round()

title_metacritic = (title_averages["metacritic"].mean()).round()

title_stats = pd.DataFrame({"Worldwide Gross Sales": title_gross,
                             "Average IMDB Rating": title_imdb_rating,
                             "Average Metacritic": title_metacritic})

title_stats
```

```python
top_title_stats = title_stats.loc[title_stats["Worldwide Gross Sales"] > 1062, [
    "Worldwide Gross Sales", "Average IMDB Rating", "Average Metacritic"]]

top_title_stats
```

```python
# bar chart for average gross sales by title

x_axis = np.arange(len(top_title_stats))
tick_locations = [value for value in x_axis]

plt.figure(figsize=(20,3))
plt.bar(x_axis, top_title_stats["Worldwide Gross Sales"], color='b', alpha=0.5, align="center")
plt.xticks(tick_locations, ["Avatar", "The Avengers", "Harry Potter and the Deathly Hallows - Part 2", "Frozen", "Iron Man 3", "Transformers: Dark of the Moon", "Skyfall", "Transformers: Age of Extinction", "The Dark Knight Rises", "Toy Story 3"], rotation="vertical")
plt.title("Gross Sales per Movie")
plt.ylabel("Gross Sales")
# Save as PNG
plt.savefig('gross_sales_per_movie.png')
```

```python
# Set up new dataframe with focused data for our rating analysis
ratings_data = movie_data[["title", "worldwide_gross", "imdb_rating", "metacritic"]].copy()
# Adjust Imdb rating to percentage
ratings_data.loc[:,"IMDB"] = ratings_data.loc[:,"imdb_rating"]*10
# Adjust worldwide gross to be in $100mils
ratings_data.loc[:,"Earnings"] = ratings_data.loc[:,"worldwide_gross"]/100_000_000
# Add average rating column
ratings_data.loc[:, "Average Rating"] = (ratings_data.loc[:,"IMDB"] + ratings_data.loc[:,"metacritic"])/2
ratings_data = ratings_data.rename(columns = {"metacritic" : "Metacritic"})
ratings_data
```

```python
#Add average rating column
ratings_data.loc[:, "Average Rating"] = (ratings_data.loc[:,"IMDB"] + ratings_data.loc[:,"Metacritic"])/2
ratings_data
```

```python
# Create a Scatter Plot for imdb_rating vs metacritic
ax = ratings_data.plot(kind="scatter", x="Earnings",y="IMDB", color="b", label="IMDB Rating")
ratings_data.plot(x="Earnings",y=("Metacritic"), color="r", kind = "scatter", label="Metacritic Rating", ax=ax)
ratings_data.plot(x="Earnings",y=("Average Rating"), color="g", kind = "scatter", label="Average Rating", ax=ax)
plt.xlim()
plt.xlabel('Worldwide Gross(100 million)')
plt.ylabel('Rating')
plt.title('Average Overall Rating vs Earnings')
plt.show()
plt.savefig('avg_rating_vs_earnings_scatter.png')
```

```python
bins = [300_000_000, 500_000_000, 750_000_000, 1_000_000_000, 3_000_000_000]
group_labels = ["300M to 500M", "500M to 750M", "750M to 1B", "1B to 3B"]
pd.cut(ratings_data["worldwide_gross"], bins, labels=group_labels)
```

```python
ratings_data["Gross Group"] = pd.cut(ratings_data["worldwide_gross"], bins, labels=group_labels)
ratings_data
```

```python
sns.pairplot(ratings_data, x_vars=["Earnings"], y_vars=["IMDB", "Metacritic", "Average Rating"], hue = "Gross Group",
             height=5, aspect=.8, kind="reg")
plt.savefig('avg_rating_by_earnings.png')
```

```python
# Narrow down the dataframe to only the columns we need for genre analysis.
movie_genre = movie_data[["Main_Genre", "year", "title", "worldwide_gross"]].copy()
movie_genre["worldwide_gross"] = (movie_genre["worldwide_gross"].astype(float))/1_000_000

# Rename columns for readability.
movie_genre = movie_genre.rename(columns = {"Main_Genre" : "Genre",
                                    "year" : "Year",
                                    "title" : "Title",
                                    "worldwide_gross" : "Earnings"})
# Print the new dataframe.
movie_genre
```

```python
# Create the chart.
movie_genre.groupby(["Year", "Genre"])["Earnings"].sum().unstack("Genre").plot.bar(figsize=(25, 30), width=1)
# Label our chart.
plt.xlabel('Movie Genres by Year', fontsize=20)
plt.ylabel('Movie Earnings in USD (x 1mil)', fontsize=20)
plt.title('Movie Earnings by Genre & Year (2008-2018)', fontsize=20)
plt.ylim(90, 6_000)
# Show the chart
plt.show()
# Export to PNG
plt.savefig('genre_earnings_by_year.png')
```

```python
# Create a chart to show movie success by genre overall
movie_genre.groupby(["Genre"])["Earnings"].sum().plot(kind="bar", figsize=(15, 10))
plt.xlabel('Genres', fontsize=20)
plt.ylabel('Movie Earnings in USD (x 1mil)', fontsize=20)
plt.title('Movie Earnings by Genre (2008-2018)', fontsize=20)
plt.ylim(100, 20_000)
# Show the chart
plt.show()
# Save as PNG
plt.savefig('genre_earnings_overall.png')
```
